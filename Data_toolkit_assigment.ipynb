#Q1. What is NumPy, and why is it widely used in Python?

* NumPy is a fundamental Python library used for numerical and scientific computing. It provides support for large, multi-dimensional arrays and matrices along with a collection of optimized mathematical functions. It is widely used because it offers faster computation, efficient memory usage, and vectorized operations compared to Python lists. Many libraries like Pandas, SciPy, and Scikit-learn are built on NumPy.

---

#Q2. How does broadcasting work in NumPy?

* Broadcasting is a mechanism in NumPy that allows arithmetic operations between arrays of different shapes by automatically expanding the smaller array to match the larger one without copying data. This makes array operations efficient and concise.

---
#Q3. What is a Pandas DataFrame?

* A Pandas DataFrame is a two-dimensional, size-mutable, and labeled data structure with rows and columns. It is similar to an Excel spreadsheet or SQL table and is used for data manipulation, cleaning, and analysis.

---
#Q4. Explain the use of the `groupby()` method in Pandas.

*  The `groupby()` method is used to split data into groups based on one or more columns, apply aggregation functions such as sum, mean, or count, and then combine the results. It is commonly used for data summarization and analysis.

---
#Q5. Why is Seaborn preferred for statistical visualizations?

* Seaborn is preferred because it provides high-level statistical plots with attractive default styles and easy integration with Pandas DataFrames. It simplifies complex visualizations like distributions, correlations, and categorical plots.

---
#Q6. What are the differences between NumPy arrays and Python lists?

* NumPy arrays store homogeneous data types, consume less memory, and support fast mathematical operations. Python lists can store heterogeneous data types but are slower and less memory-efficient for numerical computations.

---
#Q7. What is a heatmap, and when should it be used?

* A heatmap is a graphical representation of data where values are represented by colors. It is commonly used to visualize correlation matrices, intensity of values, and patterns in large datasets.

---
#Q8. What does the term “vectorized operation” mean in NumPy?

* Vectorized operations refer to performing operations on entire arrays without using explicit loops. These operations are faster because they are executed in optimized low-level code.

---
#Q9. How does Matplotlib differ from Plotly?

* Matplotlib is primarily used for static visualizations and is suitable for simple plots. Plotly is used for interactive visualizations that support zooming, hovering, and web-based dashboards.

---
#Q10. What is the significance of hierarchical indexing in Pandas?

* Hierarchical indexing allows multiple levels of indexing on a DataFrame or Series. It enables handling complex datasets and improves data organization and multi-dimensional analysis.

---
#Q11. What is the role of Seaborn’s pairplot() function?

* pairplot() creates pairwise relationships between numerical variables in a dataset. It is useful for exploratory data analysis to identify correlations and distributions.

---
#Q12. What is the purpose of the describe() function in Pandas?

* The describe() function provides a statistical summary of numerical data including count, mean, standard deviation, minimum, maximum, and quartiles.

---
#Q13. Why is handling missing data important in Pandas?

* Handling missing data is important because missing values can distort analysis results and lead to incorrect conclusions. Pandas provides methods to detect, remove, or fill missing values.

---
#Q14. What are the benefits of using Plotly for data visualization?

* Plotly offers interactive charts, real-time data visualization, user-friendly dashboards, and easy integration with web applications.

---
#Q15. How does NumPy handle multidimensional arrays?

* NumPy handles multidimensional arrays using ndarray objects that store data in a structured format with defined axes, shapes, and dimensions.

---
#Q16. What is the role of Bokeh in data visualization?

* Bokeh is a Python library used to create interactive and scalable visualizations for modern web browsers. It is commonly used in dashboards and web applications.

---
#Q17. Explain the difference between apply() and map() in Pandas.

* map() is used on Pandas Series for element-wise operations. apply() works on both Series and DataFrames and can apply functions row-wise or column-wise.

---
#Q18. What are some advanced features of NumPy?

* Advanced features include broadcasting, vectorization, linear algebra operations, random number generation, Fourier transforms, and memory-efficient slicing.

---
#Q19. How does Pandas simplify time series analysis?

* Pandas simplifies time series analysis by providing datetime indexing, resampling, rolling window operations, and time-based filtering.

---
#Q20. What is the role of a pivot table in Pandas?

* A pivot table summarizes large datasets by aggregating values based on categories. It is used to analyze trends and patterns efficiently.

---
#Q21. Why is NumPy’s array slicing faster than Python’s list slicing?

* NumPy slicing returns views instead of copies, which reduces memory usage and increases performance compared to Python list slicing.

---
Q22. What are some common use cases for Seaborn?
-

* Seaborn is used for exploratory data analysis, correlation analysis, distribution visualization, regression analysis, and categorical data visualization.

```python
#1. How do you create a 2D NumPy array and calculate the sum of each row?


import numpy as np

arr = np.array([[1, 2, 3],
                [4, 5, 6]])

row_sum = arr.sum(axis=1)
print(row_sum)
```

```python
#2. Write a Pandas script to find the mean of a specific column in a DataFrame.

import pandas as pd

df = pd.DataFrame({
    "Marks": [70, 80, 90, 85]
})

mean_value = df["Marks"].mean()
print(mean_value)
```

```python
#3. Create a scatter plot using Matplotlib.

import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.scatter(x, y)
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

```python
# 4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = sns.load_dataset("iris")
corr_matrix = df.drop('species', axis=1).corr()
sns.heatmap(corr_matrix, annot=True)
plt.show()
```

```python
# 5. Generate a bar plot using Plotly.

import plotly.express as px
import pandas as pd # Import pandas

data = {
    "Product": ["A", "B", "C"],
    "Sales": [200, 150, 300]
}

df = pd.DataFrame(data) # Correctly create a Pandas DataFrame
fig = px.bar(df, x="Product", y="Sales")
fig.show()
```

```python
#6. Create a DataFrame and add a new column based on an existing column.

import pandas as pd

df = pd.DataFrame({
    "Price": [100, 200, 300]
})

df["Discounted_Price"] = df["Price"] * 0.9
print(df)
```

```python
#7. Write a program to perform element-wise multiplication of two NumPy arrays.

import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

result = a * b
print(result)
```

```python
#8. Create a line plot with multiple lines using Matplotlib.

import matplotlib.pyplot as plt

x = [1, 2, 3, 4]

plt.plot(x, [1, 4, 9, 16], label="Square")
plt.plot(x, [1, 8, 27, 64], label="Cube")

plt.legend()
plt.show()
```

```python
# 9. Generate a Pandas DataFrame and filter rows where a column value is
     #greater than a threshold.

import pandas as pd

df = pd.DataFrame({
    "Score": [45, 60, 75, 30]
})

filtered_df = df[df["Score"] > 50]
print(filtered_df)
```

```python
#11. Perform matrix multiplication using NumPy.

import numpy as np

A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

result = np.dot(A, B)
print(result)
```
